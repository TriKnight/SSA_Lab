```python
import numpy as np
import matplotlib.pyplot as plt
from pylab import meshgrid,cm,imshow,contour,clabel,colorbar,axis,title,show
#Define the function in 3D
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
```

```python
# Random Search for maximize problem
def z_func(x1,x2):
    return (3*(1-x1)**2*np.exp(-x1**2-(x2+1)**2)
            +10*(x1/5 -x1**3-x1**5)*np.exp(-x1**2-x2**2) 
            -1/3*np.exp(-(x1+1)**2-x2**2))
# range from -3 to 3 for each x1 and x2
x1 = np.arange(-3.0,3.0,0.1)
x2 = np.arange(-3.0,3.0,0.1)
X,Y = meshgrid(x1, x2) # grid of point
Z = z_func(X, Y) # evaluation of the function on the grid
im = imshow(Z,cmap=cm.RdBu) # drawing the function
# latex fashion title
colorbar(im) # adding the colobar on the right
title('Equation 2')
show()

#Define the function in 3D
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, 
                      cmap=cm.RdBu,linewidth=0, antialiased=False)

ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))
ax.view_init(elev=30., azim=-25, roll=0)
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```
