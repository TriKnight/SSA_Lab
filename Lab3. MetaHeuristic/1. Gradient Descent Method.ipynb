# 1. Gradient Descent

Minimize the function $f(x_1, x_2) = 10-e^{(-x_1^2 + 3\times x_2^2)}$ 

subjec to $-1 < \leq x_1 \leq 1$ and $-1 < \leq x_2 \leq 1$

### Alogithm Gradient Descent Method 
1. $ k \leftarrow 0 $
2. $ x_1^k \leftarrow Random [-1, 1],  x_2^k \leftarrow Random [-1, 1]$
3. **While** $k$ < $N inter${

```python
#Gradient Descent
import numpy as np
import matplotlib.pyplot as plt
from pylab import meshgrid,cm,imshow,contour,clabel,colorbar,axis,title,show
```

```python
# Define a function in 2D
def z_func(x1,x2):
    return (10-np.exp(-(x1**2+3*x2**2)))
x1 = np.arange(-1.0,1.0,0.1)
x2 = np.arange(-1.0,1.0,0.1)
X,Y = meshgrid(x1, x2) # grid of point
Z = z_func(X, Y) # evaluation of the function on the grid
im = imshow(Z,cmap=cm.RdBu) # drawing the function
# latex fashion title
colorbar(im) # adding the colobar on the right
title('$z=10-exp(-(x1**2+3*x2**2))$')
show()
```

```python
#Define the function in 3D
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, 
                      cmap=cm.RdBu,linewidth=0, antialiased=False)

ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```

```python
## Optimization process
# Define number of interations
k=0;
n_iter =200;
# gradient step size h difinition
hstep =0.001;
#Step size of the Gradient Descent method
alpha = 0.05;
# Random starting point
x1_start = 0.8
x2_start = 0.8
# x1 and x2 array
x1_array=[]
x2_array=[]
zn_array=[]
def gradient_descent(x1, x2, n_iter, alpha):
    #Compute the gradient gx1 and gx2
    for i in range(n_iter):
        zn = z_func(x1, x2)
        vx1= x1 + hstep
        vx2 = x2+ hstep
        gx1 = (z_func(vx1, x2)-zn)/hstep
        gx2 = (z_func(x1, vx2)-zn)/hstep

        # Computation of the new solution
        x1 = x1 - alpha*gx1
        x2 = x2 - alpha*gx2
        x1_array.append(x1)
        x2_array.append(x2)
        zn_array.append(zn)
gradient_descent(x1_start, x2_start, n_iter, alpha)
```

```python
#Define the function in 3D
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.plot_wireframe(X, Y, Z, rstride=1, cstride=1)
ax.plot(x1_array, x2_array, zn_array, zdir='z', label='curve in (x, y)',
        marker='o', linestyle='dashed',linewidth=2, markersize=6, color='r')

ax.view_init(elev=30., azim=-35, roll=0)
ax.set_xlim(-1, 1)
ax.set_ylim(-1, 1)
ax.set_zlim(9, 10)



plt.show()
```
