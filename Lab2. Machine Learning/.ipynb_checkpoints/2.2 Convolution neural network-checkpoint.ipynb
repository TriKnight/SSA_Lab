##  Improving Fashion MNIST dataset with CNN 

```python
import tensorflow as tf

# Load the Fashion MNIST dataset
fmnist = tf.keras.datasets.fashion_mnist
(training_images, training_labels), (test_images, test_labels) = fmnist.load_data()

# Normalize the pixel values
training_images = training_images / 255.0
test_images = test_images / 255.0
```

```python
model = tf.keras.models.Sequential([
	# With 64 filters and kernel size 3x3.
	tf.keras.layers.Conv2D(64, (3,3), activation='relu', input_shape=(28,28,1)),
	tf.keras.layers.MaxPooling2D(2,2),
	tf.keras.layers.Conv2D(64,(3,3),activation='relu'),
	tf.keras.layers.MaxPooling2D(2,2),
	tf.keras.layers.Flatten(),
	tf.keras.layers.Dense(128, activation='relu'),
	tf.keras.layers.Dense(10, activation='softmax') 
])
#Show the detail the model
model.summary()
```

```python
# Define the model
model = tf.keras.models.Sequential([
                                                         
  # Add convolutions and max pooling
  tf.keras.layers.Conv2D(32, (3,3), activation='relu', input_shape=(28, 28, 1)),
  tf.keras.layers.MaxPooling2D(2, 2),
  tf.keras.layers.Conv2D(32, (3,3), activation='relu'),
  tf.keras.layers.MaxPooling2D(2,2),

  # Add the same layers as before
  tf.keras.layers.Flatten(),
  tf.keras.layers.Dense(128, activation='relu'),
  tf.keras.layers.Dense(10, activation='softmax')
])

# Print the model summary
model.summary()

# Use same settings
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the model
print(f'\nMODEL TRAINING:')
model.fit(training_images, training_labels, epochs=5)

# Evaluate on the test set
print(f'\nMODEL EVALUATION:')
test_loss = model.evaluate(test_images, test_labels)
```

### The result by using CNN (92-93%) more accuracy than Dense NN (86-87%). And the number of parameters more less

## Visualizing the  Convolution and Pooling
The print (test_labels[:100]) shows us the first 100 labels in the test set, 
and you can see that the ones at index 0, index 23 and index 28 are all the same value (9). They're all shoes

```python
print(test_labels[:100])
```

```python
import matplotlib.pyplot as plt
f, axarr = plt.subplots(3,4)
FIRST_IMAGE=0 # First row of image
SECOND_IMAGE=23 # Second row of image
THIRD_IMAGE=28  ## Third row of image


CONVOLUTION_NUMBER = 20 # The order number of Conv (in our case 0-63 Conv layer)
from tensorflow.keras import models
layer_outputs = [layer.output for layer in model.layers]
# 4 activation layers: Conv2D -> Pooling --> Conv2D --> Pooling 
activation_model = tf.keras.models.Model(inputs = model.input, outputs = layer_outputs)
for x in range(0,4):
    f1 = activation_model.predict(test_images[FIRST_IMAGE].reshape(1, 28, 28, 1))[x]
    axarr[0,x].imshow(f1[0, : , :, CONVOLUTION_NUMBER], cmap='inferno')
    axarr[0,x].grid(False)
    f2 = activation_model.predict(test_images[SECOND_IMAGE].reshape(1, 28, 28, 1))[x]
    axarr[1,x].imshow(f2[0, : , :, CONVOLUTION_NUMBER], cmap='inferno')
    axarr[1,x].grid(False)
    f3 = activation_model.predict(test_images[THIRD_IMAGE].reshape(1, 28, 28, 1))[x]
    axarr[2,x].imshow(f3[0, : , :, CONVOLUTION_NUMBER], cmap='inferno')
    axarr[2,x].grid(False)
```

To visualizing the conv2D and Pooling, we can use this code bellow. It's can show each convolution filter by using CONVOLUTION_NUMBER parameter. We can change this from [0-63] and see the machine learn very specific location in the images depends on kind of filter.
