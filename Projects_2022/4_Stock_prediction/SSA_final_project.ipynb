```python
# Load train and test data set
train=pd.read_csv(r'/content/gdrive/MyDrive/SSA_lab/train.txt', header=None)
test=pd.read_csv(r'/content/gdrive/MyDrive/SSA_lab/test.txt', header=None)
```

```python
# Take the first 50 collumns for features
xtrain=train.iloc[:,0:50]
xtrain=np.array(xtrain)
xtrain=np.reshape(xtrain,(800,50))
```

```python
# Take the last collumn for ground truths
ytrain=train.iloc[:,50:51]
ytrain=np.array(ytrain)
ytrain=np.reshape(ytrain,(800,1))
```

```python
# Take the first 50 collumns for features
xtest=test.iloc[:,0:50]
xtest=np.array(xtest)
xtest=np.reshape(xtest,(200,50))
```

```python
# Take the last collumn for ground truths
ytest=test.iloc[:,50:51]
ytest=np.array(ytest)
ytest=np.reshape(ytest,(200,1))
```

```python
import tensorflow as tf 
import numpy as np
from tensorflow import keras
print(tf.__version__)

# Model with 1 convolution layer and 1 Dense Net layer
model = keras.models.Sequential([
    keras.layers.Conv1D(filters = 32, 
                               kernel_size = 2,
                               input_shape = (50,1), 
                               activation = "relu"),
    keras.layers.MaxPooling1D(2),
    keras.layers.Flatten(),
    keras.layers.Dense(2, activation='softmax')
    ])

model.compile(optimizer='adam', loss='mse', metrics=['accuracy'])
model.fit(xtrain, ytrain, epochs=1000)

# Print the prediction of the result
print(model.predict([[32.5335,33.112,33.044,33.078,33.078,32.7377,32.2613,32.5335,32.2613,32.4995,31.2403,29.743,31.1042,30.3215,30.7299,30.2875,31.2403,31.6827,31.989,31.8529,32.0571,32.1251,31.989,32.3293,32.6356,32.5335,32.6736,32.5693,32.5693,32.7083,32.9517,33.6816,33.9597,33.6468,34.6549,34.6896,34.4115,34.6896,35.0372,34.7591,33.7164,34.4115,34.203,34.203,34.1682,34.2725,34.7591,35.1067,34.7591,35.5934]]))
```

```python
# Produce validation result
test_loss = model.evaluate(xtest, ytest)
```
